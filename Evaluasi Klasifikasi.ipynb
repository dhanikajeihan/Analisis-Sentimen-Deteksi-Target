```python
import pandas as pd
```

```python
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("Hasil Analisis Sentimen.csv")
df.drop(columns=['normalisasi', 'clean', 'token', 'likelihood positif (bow)', 'likelihood negatif (bow)', 'likelihood positif (lb)', 'likelihood negatif (lb)'])
```

# Evaluasi Bag of Words

```python
cm_bow = confusion_matrix(df["kelas aktual"],df["kelas prediksi (bow)"], labels=["positif","negatif"])
cm_bow
```

```python
print("Bag of Words\n")
precision_bow = cm_bow[0][0]/(cm_bow[0][0]+cm_bow[0][1])
print("Precision", precision_bow)
recall_bow = cm_bow[0][0]/(cm_bow[0][0]+cm_bow[1][0])
print("Recall", recall_bow)
fm_bow = (2*precision_bow*recall_bow)/(precision_bow+recall_bow)
print("F-Measure", fm_bow)
akurasi_bow = (cm_bow[0][0]+cm_bow[1][1])/(cm_bow[0][0]+cm_bow[0][1]+cm_bow[1][0]+cm_bow[1][1])
print("Akurasi", akurasi_bow)
```

# Evaluasi Lexicon-based

```python
cm_lb = confusion_matrix(df["kelas aktual"],df["kelas prediksi (lb)"], labels=["positif","negatif"])
cm_lb
```

```python
print("Lexicon-based\n")
precision_lb = cm_lb[0][0]/(cm_lb[0][0]+cm_lb[0][1])
print("Precision", precision_lb)
recall_lb = cm_lb[0][0]/(cm_lb[0][0]+cm_lb[1][0])
print("Recall", recall_lb)
fm_lb = (2*precision_lb*recall_lb)/(precision_lb+recall_lb)
print("F-Measure", fm_lb)
akurasi_lb = (cm_lb[0][0]+cm_lb[1][1])/(cm_lb[0][0]+cm_lb[0][1]+cm_lb[1][0]+cm_lb[1][1])
print("Akurasi", akurasi_lb)
```

# Evaluasi Gabungan Bag of Words dan Lexicon-based

```python
cm_gabungan = confusion_matrix(df["kelas aktual"],df["kelas prediksi fitur gabungan"], labels=["positif","negatif"])
cm_gabungan
```

```python
print("Fitur BoW dan Lexicon-based\n")
precision_gabungan = cm_gabungan[0][0]/(cm_gabungan[0][0]+cm_gabungan[0][1])
print("Precision", precision_gabungan)
recall_gabungan = cm_gabungan[0][0]/(cm_gabungan[0][0]+cm_gabungan[1][0])
print("Recall", recall_gabungan)
fm_gabungan = (2*precision_gabungan*recall_gabungan)/(precision_gabungan+recall_gabungan)
print("F-Measure", fm_gabungan)
akurasi_gabungan = (cm_gabungan[0][0]+cm_gabungan[1][1])/(cm_gabungan[0][0]+cm_gabungan[0][1]+cm_gabungan[1][0]+cm_gabungan[1][1])
print("Akurasi", akurasi_gabungan)
```

# Evaluasi Kelas Positif

# Evaluasi Kelas Negatif

```python
dfn = pd.read_csv("Hasil Analisis Sentimen - Neg.csv")
dfn.drop(columns=['normalisasi', 'clean', 'token', 'likelihood positif (bow)', 'likelihood negatif (bow)', 'likelihood positif (lb)', 'likelihood negatif (lb)'])
```

```python
cm_gabungan_neg = confusion_matrix(dfn["kelas aktual"],dfn["kelas prediksi fitur gabungan"], labels=["negatif","positif"])
cm_gabungan_neg
```

```python
print("Fitur Gabungan (Kelas Negatif)\n")
precision_gabungan_neg = cm_gabungan_neg[0][0]/(cm_gabungan_neg[0][0]+cm_gabungan_neg[0][1])
print("Precision", precision_gabungan_neg)
recall_gabungan_neg = cm_gabungan_neg[0][0]/(cm_gabungan_neg[0][0]+cm_gabungan_neg[1][0])
print("Recall", recall_gabungan_neg)
fm_gabungan_neg = (2*precision_gabungan_neg*recall_gabungan_neg)/(precision_gabungan_neg+recall_gabungan_neg)
print("F-Measure", fm_gabungan_neg)
akurasi_gabungan_neg = (cm_gabungan_neg[0][0]+cm_gabungan_neg[1][1])/(cm_gabungan_neg[0][0]+cm_gabungan_neg[0][1]+cm_gabungan_neg[1][0]+cm_gabungan_neg[1][1])
print("Akurasi", akurasi_gabungan_neg)
```

# Evaluasi Macro Averaging

```python
print("Macro Averaging Fitur BoW dan Lexicon-based \n")
macro_precision_gabungan = (precision_gabungan_pos+precision_gabungan_neg)/2
print("Precision", macro_precision_gabungan)
macro_recall_gabungan = (recall_gabungan_pos+recall_gabungan_neg)/2
print("Recall", macro_recall_gabungan)
macro_fm_gabungan = (fm_gabungan_pos+fm_gabungan_neg)/2
print("F-Measure", macro_fm_gabungan)
macro_akurasi_gabungan = (akurasi_gabungan_pos+akurasi_gabungan_neg)/2
print("Akurasi", macro_akurasi_gabungan)
```
